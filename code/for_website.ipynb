```python
# import necessary libraries
import os
import warnings
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# import csv file
model1_return = pd.read_csv("../prediction/cnn_model1_seed0_epochs100_days5_stride1_diffFalse_good.csv")
model2_return = pd.read_csv("../prediction/cnn_model3_seed6_epochs100_days5_stride1_diffFalse_good.csv")
model3_return = pd.read_csv("../prediction/cnn_model3_seed11_epochs100_days5_stride1_diffFalse_good.csv")
```

```python
close = pd.read_csv('../data/data.csv', index_col=0)['close']
y_true = close.iloc[-276:-6]
```

```python
warnings.filterwarnings("ignore")

y_pred1 = []
y_pred1.append(y_true.iloc[0])

i = 1
while i < len(model1_return)-1:
    y_pred1.append(y_pred1[i-1]*(1+float(model1_return.iloc[i-1].values)))
    i += 1

y_pred2 = []
y_pred2.append(y_true.iloc[0])

i = 1
while i < len(model2_return)-1:
    y_pred2.append(y_pred2[i-1]*(1+float(model2_return.iloc[i-1].values)))
    i += 1
    
y_pred3 = []
y_pred3.append(y_true.iloc[0])

i = 1
while i < len(model3_return)-1:
    y_pred3.append(y_pred3[i-1]*(1+float(model3_return.iloc[i-1].values)))
    i += 1
```

```python
plt.figure(figsize=(12, 4))
plt.plot(y_true, label='true', color='#4a8cff')
plt.plot(y_pred1, label='model1', color='#ff9a9a', linestyle='--')
plt.plot(y_pred2, label='model2', color='#d68d2e', linestyle='--')
plt.plot(y_pred3, label='model3', color='#b7d1ff', linestyle='--')

plt.legend()

plt.gca().spines['left'].set_color('#d9d9d9')
plt.gca().spines['bottom'].set_color('#d9d9d9')
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)
plt.tick_params(axis='x', colors='#d9d9d9')
plt.tick_params(axis='y', colors='#d9d9d9')
plt.xticks(color='#595959')
plt.yticks(color='#595959')
plt.xlabel('Date', color='#595959')
plt.ylabel('Price', color='#595959')
plt.legend(loc='upper right', frameon=False, facecolor='None', edgecolor='None', labelcolor='#595959', ncol = 2)
plt.gca().patch.set_facecolor('None')
plt.gca().get_legend().get_frame().set_linewidth(0.0)
plt.gca().xaxis.set_major_locator(plt.MultipleLocator(30))
plt.xticks(y_true.index[::int(len(y_pred1) / 5)])

plt.savefig('./price.png', transparent=True)
plt.show()
```
